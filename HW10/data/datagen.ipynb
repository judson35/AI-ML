```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150
```

```python
def dataGen():
    # Set random seed so generated random numbers are always the same
    gen = np.random.RandomState(7)
    # Length of generated data
    n = 200
    # Generate x and y
    x = gen.rand(n)
    y = np.sin(4*np.pi*x) + 1*np.exp(2*x)
    # Add some noise to most
    idx1 = gen.choice(np.arange(n), int(n*0.75), replace = False)
    y[idx1] = y[idx1] + 0.5*(gen.rand(len(idx1)) - 0.5)
    # Add more noise to the rest
    idx2 = list(set(np.arange(n)) - set(idx1))
    y[idx2] = y[idx2] + 2.5*(gen.rand(len(idx2)) - 0.5)
    # Return x and y
    return x, y, idx1, idx2

x, y, idx1, idx2 = dataGen()

plt.scatter(x[idx1],y[idx1], c = 'blue', alpha = 0.5)
plt.scatter(x[idx2],y[idx2], c = 'green', alpha = 0.5)
plt.show()
```

```python
np.savetxt('m10-hw1-data.txt', np.vstack((x,y)).T)
```
